#### Cleaning dates: changing formats from Jan-15 to 2015/01

```python
month_dict = {"Jan":'01', "Feb":'02', "Mar":'03', "Apr":'04', 
              "May":'05', "Jun":'06', "Jul":'07', "Aug":'08', 
              "Sep":'09', "Oct":'10', "Nov":'11', "Dec":'12', }
```

```python
def fixdate(d):
    if type(d) is str:
        d=d.split("-")
        if int(d[1]) > 20:
            yr = "19" + d[1]
        else:
            yr = "20" + d[1]
        d = yr + "/" + month_dict[d[0]]
        return d
    return float('NaN')
```

```python
df.last_pymnt_d=df.last_pymnt_d.apply(lambda x: fixdate(x))
df.next_pymnt_d=df.next_pymnt_d.apply(lambda x: fixdate(x))
df.last_credit_pull_d=df.last_credit_pull_d.apply(lambda x: fixdate(x))
df.issue_d=df.issue_d.apply(lambda x: fixdate(x))
df.earliest_cr_line=df.earliest_cr_line.apply(lambda x: fixdate(x))
```

Let's now segment the emp_length into 4 segments 
- 0-3 Junior
- 3-8 MidLevel
- 8-10 Senior

```python
def emp_bucket(value):
    if value in range(0, 4):
        return 'Junior'
    if value in range(4, 9):
        return "Mid"
    
    return "Senior"

dataset['emp_length_segment']=dataset.emp_length.apply(lambda x: emp_bucket(x))
```

#### Business driven metrics 
- Lets breakup dti into low, medium, high  
- Lets create a raito of annual_income to loan_amount
- Lets create a ratio of annual_income to installment

```python
def dti_breakup(val):
    if val <= 10:
        return "low"
    if val <= 20:
        return "medium"
    else:
        return "high"
dataset['dti_segments'] = dataset.dti.apply(lambda x: dti_breakup(x))
```

```python
dataset['income_loan_amount_ratio']=round((dataset.annual_inc / dataset.loan_amnt), 2)
```

```python
dataset['income_installment_ratio']=round((dataset.annual_inc / dataset.installment), 2)
```

```python
dti_segments_status = dataset.groupby('loan_status')['dti_segments'].value_counts(ascending=False).unstack(level=0).reset_index()
dti_segments_status.fillna(0, inplace=True)
dti_segments_status['Total'] = dti_segments_status['Charged Off'] + dti_segments_status['Fully Paid']
dti_segments_status['Charged_Off_Percent'] = dti_segments_status['Charged Off'] * 100 /dti_segments_status['Total']
dti_segments_status['Fully_Paid_Percent'] = dti_segments_status['Fully Paid'] * 100 /dti_segments_status['Total']
dti_segments_status.sort_values(by='Charged_Off_Percent', ascending=False, inplace=True)

dti_segments_status
```

```python
fig = plt.figure(figsize=(20, 5), )

plt.subplot(1, 3, 1)
ax=sns.countplot(x="dti_segments", hue="loan_status", data=dataset, linewidth=1, order=['high', 'medium', 'low'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.title("Counts of Lona-Status vs DTI-Segments")
plt.ylabel("# of Loan Applications")
for p in ax.patches:
    h = round(p.get_height(),2)
    x,y = p.get_xy()
    x  += .2
    y =  h + 0.25
    ax.annotate('{}'.format(h), (x,y))

    
plt.subplot(1, 3, 2)
ax=sns.countplot(x="dti_segments", data=dataset[dataset.loan_status=="Charged Off"], linewidth=1, order=['high', 'medium', 'low'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.title("Counts of Lona-Status vs DTI-Segments")
plt.ylabel("# of Loan Applications")
for p in ax.patches:
    h = round(p.get_height(),2)
    x,y = p.get_xy()
    x  += .1
    y =  h + 20
    ax.annotate('{}'.format(h), (x,y))
    

plt.subplot(1, 3, 3)
ax=sns.barplot(x="dti_segments", y="Charged_Off_Percent", data=dti_segments_status)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.title("%ages of Lona-Status vs DTI-Segments")
plt.ylabel('Charged Off(%)')
for p in ax.patches:
    h = round(p.get_height(),2)
    x,y = p.get_xy()
    x  += .2
    y =  h + 0.25
    ax.annotate('{}'.format(h), (x,y))

    
plt.show()
```

## Recommendations


- Thorough scrutiny should be done on applicants citing Small Business as a purpose for the Loan.
- Furthermore, High Loan amount should not be sanctioned for people giving Small Business or Debt Consolidation as a purpose for the Loan
- Either a thorough background check on the applicant should be done or interest charged should be high enough to discourage such Loans.
- Applicants with High Debt to Income ratio (DTI) should be treated with caution. Thorough scrutiny and background check must be done. 
- Applicants with High DTI ratio citing Credit Card or Debt Consolidation as a purpose must be strictly dealt with, security mortgaging must be done if need be since they are Highly likely to Default.
- Interest Rates should be reduced for people citing House as a purpose for Loan. Since High interest rate here implies high default rate
- Caution should taken when Applicants having Home Ownership as Others and are willing to take a Loan at a High Interest Rate. Either a security mortgaging should be taken from such applicant or a thourough background check must be done. Since they are very likely to default. 
- Caution should be maintained when giving loans to applicants with High Annual Income BUT having Mortgage as Home Ownership. Since they are very Likely to default.
- There seems some loopholes in the verification check as we can see verified people have high %age of Charged-off
